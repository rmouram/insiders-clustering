### Task - Tarefas

1. Quem são as pessoas elegíveis ?
    - O que é ser elegível? O que são clientes de maior "valor"
    - Faturamento:
        - Alto ticket médio.
        - Alto LTV.
        - Baixo Recência.
        - Alto basket size.
        - Baixa probabilidade de churn.
        - Alta propensão de compra.
    - Custo:
        - Baixa taxa de devolução.
    - Experiencia de compra:
        - Média alta das avaliações
2. Quantos clientes farão parte do grupo?
    - Número total de clientes.
    - % do grupo Insiders.
3. Quais as características desses clientes?
    - Escrever características:
        - Idade.
        - Localização.
        - Gênero.
    - Características do consumo:
        - Features da clusterização.
4. Qual a porcentagem de contribuição do faturamento, vindo do insiders?
    - Faturamento total do ano.
    - Faturamento do grupo Insiders.
5. Qual a expectativa de faturamento desse grupo para os proximos meses ?
    - LTV do grupo Insiders.
    - Análise de Cohort.
6. Quais as condições para uma pessoa ser elegível?
    - Definir peridodicidade.
    - A pessoa precisa ser similar ou parecido com uma pessoa grupo.
7. Quais as condições para uma pessoa ser removida?
    - A pessoa precisa ser dissimilar ou não-parecido com uma pessoa do grupo.
8. Qual a garantia que o programa Insiders é melhor que o resto da base?
    - Teste A/B.
    - Teste A/B Baysiano.
    - Teste de hipóteses.
9. Quais ações o time de marketing pode realizar para aumentar o faturamento?
    - Descontos.
    - Preferência de compra.
    - Frete.
    - Itens exclusivos.

#### RFM Model

Para modelos de negócio de e-commerce com recência

1. Recência  
    a. Tempo desde a última compra.  
    b. Responsividade.  

2. Frequência  
    a. Quantos produtos compra em um período.  
    b. Engajamento.  
    
3. Monetário  
    a. Total gasto, faturamento.  
    b. "High-value purchases"  
    

##### RFM Score

1. Ordenar clientes por recência
2. Dar notas aos clientes
    i. notas de 1-5(5 notas)
    ii. 5 grupos de 20% cada
3. Fazer o mesmo para frequência e monetário
4. Calcular média das notas para R, F, M  
    Exemplo: Customer ID | R,F,M notas | Média
                1            5,4,4         4.3
                2            1,1,1         1
                3            4,3,3         3,3

##### RFM Segmentation

1. Champions  
    a. Compras recentes, frequentes com alto valor gasto  
    b. Prêmios para esses clientes  
2. Potential Loyalists  
    a. Compras recentes, boa frequencia e bom valor gasto   
    b. Programa de fidelização e Upsell  
3. New Customers  
    a. Compra recente, baixa frequência  
    b. Cosntrução de relacionamento, ofertas especiais  
4. At Risk Customer  
    a. Faz tempo que não compra  
    b. Campanhas de reativação, ofertas, produtos  
5. Can't Lose Them
